The main idea in preprocessing the data was to **enlarge** the training dataset in order to make the training more robust. This was done in several steps:

+ Randomly rotate images
+ Add random Gaussian noise to images

The new training set then consits of four times the original size:

1. Original images
2. Rotated images
3. Original images with random noise
4. Rotated images with random noise

The train/valdiation splits are *not* perfomed here anymore! Why? Of course, a training and validation splitting was used during the definition and fine-tuning of the model (see the code in the appendix for my definition of train-validation splitting).
````python
### Generate data additional data (OPTIONAL!)
### and split the data into training/validation/testing sets here.
### Feel free to use as many code cells as needed.
import cv2
LOWER_ANGLE = -15
UPPER_ANGLE = 15

def rotate_image(img, rotation_angle):
    rows, cols = img.shape[0:2]
    

    M = cv2.getRotationMatrix2D((cols/2,rows/2), rotation_angle, 1)
    dst = cv2.warpAffine(img,M,(cols,rows))
    dst = np.reshape(dst, (32, 32, 3))
    
    return(dst)

def rotate_all_images(images, LOWER_ANGLE, UPPER_ANGLE):
    return np.array([rotate_image(img, np.random.randint(LOWER_ANGLE, UPPER_ANGLE)) for img in images])

X_train = np.vstack([X_train, rotate_all_images(X_train, LOWER_ANGLE, UPPER_ANGLE)])
y_train = np.hstack([y_train, y_train])

MAX_JITTER = 20

def jitter_image(img, max_jitter):
    h,w,c = img.shape 
    
    noise = np.random.randint(0, max_jitter, (h, w, c)) # design jitter/noise here
    zitter = np.zeros_like(img)
    
    zitter[:, :, 0] = noise[:, :, 0]
    zitter[:, :, 1] = noise[:, :, 1]
    zitter[:, :, 2] = noise[:, :, 2]
    noise_added = cv2.add(img, zitter)
    
    return noise_added

def jitter_all_images(images, max_jitter):
    jittered_list = [jitter_image(img, MAX_JITTER) for img in images]
    return(np.array(jittered_list))

X_train = np.vstack([X_train, jitter_all_images(X_train, MAX_JITTER)])
y_train = np.hstack([y_train, y_train])
````

````python
import cv2

def equalize_hist(img):
    img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)

    # equalize the histogram of the Y channel
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    img_yuv[:,:,1] = cv2.equalizeHist(img_yuv[:,:,1])
    img_yuv[:,:,2] = cv2.equalizeHist(img_yuv[:,:,2])

    # convert the YUV image back to RGB format
    img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

    return(img_output)

def equalize_hist_all(images):
    return(np.array([equalize_hist(img) for img in images]))

X_train_hist_eq = equalize_hist_all(X_train)
#X_validation_hist_eq = equalize_hist_all(X_validation)
X_test_hist_eq = equalize_hist_all(X_test)
````

### Question 3

_What does your final architecture look like? (Type of model, layers, sizes, connectivity, etc.)  For reference on how to build a deep neural network using TensorFlow, see [Deep Neural Network in TensorFlow
](https://classroom.udacity.com/nanodegrees/nd013/parts/fbf77062-5703-404e-b60c-95b78b2f3f9e/modules/6df7ae49-c61c-4bb2-a23e-6527e69209ec/lessons/b516a270-8600-4f93-a0a3-20dfeabe5da6/concepts/83a3a2a2-a9bd-4b7b-95b0-eb924ab14432) from the classroom._


**Answer:** The final architecture looks very similar to *LeNet*. The activation functions are always RELUs.  It has the following layers:

1. Convolutional layer with a 5x5 filter, a stride of 1 and a depth of 16. 
2. Max-Pooling layer of size 2x2 and a stride of 2.
3. Convolutional layer with a 5x5 filter, a stride of 1 and a depth of 32. 
4. Max-Pooling layer of size 2x2 and a stride of2.
5. Fully connected layer with 200 hidden units and a dropout.
6. Fully connected layer with 200 hidden units and a dropout.

This was the architecture on which a single CNN was trained. For each differently preprocessed set of images, five neural networks with different sets of learning parameters were trained. The final logits were the average of the 25 predictions (five differently preprocessed sets, each of which was trained five times). Instead of taking the mean and thus weighting each neural net the same, it would have been possible to train a final logistic unit. This step was not performed, since according to Ref [2], in the example of recognizing handwritten digits, it was preferable to take the average (the reason is probably that another training set is required to train the final layer thus making the individual nets less powerful).

If the errors of the different neural networks are uncorrelated, then this bagging operation is expected to drastically reduce the error. The different preprocessing techniques in combination with the different learning parameters (and also the different initial weights) are supposed to make sure that the errors are not too correlated. The bagging worked reasonably well: the best cross validation result for a single architecture (since we pick the best, this probably overestimates the accuracy) suggested an accuracy of roughly 96.5%, the bagged network had 98.3%, i.e. a performance improvement of 1.8% and almost reducing the error rate by (relative) 50%.

````python
from tensorflow.contrib.layers import flatten

{'conv_filter_size_1': [5, 5, 3, 6], 
                'conv_filter_size_2': [5, 5, 6, 16],
                'fc1_size': 300}



def train_single_dnn(x, params, MU, SIGMA, DROPOUT_RATE):
    """
    Trains a single deep neural network to be
    used in the formation of the MCDNN. The 
    architecture of a single DNN is
    1. convolutional layer
    2. max pooling layer
    3. convolutional layer
    4. max pooling layer
    5. fully connected layer
    6. fully connected layer
    
    The parameters of layers can be set with the
    param argument.
    
    :param x: training data, a 32x32x3 image
    :params params: a dictionary containing all parameters needed for training
    """
    # First convolutional layer
    conv1_W = tf.Variable(tf.truncated_normal(params['conv_filter_size_1'], mean = MU, stddev = SIGMA), name='weight_1')
    conv1_b = tf.Variable(tf.zeros(params['conv_filter_size_1'][3]), name='bias_1')
    conv1 = tf.nn.conv2d(x, conv1_W, strides = [1, 1, 1, 1], padding='SAME') + conv1_b
    conv1 = tf.nn.relu(conv1)
    
    # Max pooling layer
    conv1 = tf.nn.max_pool(conv1, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
    
    # Second convolutional layer
    conv2_W = tf.Variable(tf.truncated_normal(params['conv_filter_size_2'], mean = MU, stddev = SIGMA), name='weight_2')
    conv2_b = tf.Variable(tf.zeros(params['conv_filter_size_2'][3]), name='bias_2')
    conv2 = tf.nn.conv2d(conv1, conv2_W, strides = [1, 1, 1, 1], padding='SAME') + conv2_b
    conv2 = tf.nn.relu(conv2)
    
    # Max pooling layer
    conv2 = tf.nn.max_pool(conv2, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
    
    # Flatten last convolutional layer
    fc0 = flatten(conv2)
    fc1_shape = int(fc0.get_shape()[1])
    
    # First fully connected layer
    fc1_W = tf.Variable(tf.truncated_normal(shape=(fc1_shape, params['fc1_size']), mean=MU, stddev=SIGMA), name ='weight_3')
    fc1_b = tf.Variable(tf.zeros(params['fc1_size']), name='bias_3')
    fc1 = tf.matmul(fc0, fc1_W) + fc1_b
    fc1 = tf.nn.relu(fc1)
    fc1 = tf.nn.dropout(fc1, keep_prob=DROPOUT_RATE)
    
    # Second fully connected layer
    fc2_W = tf.Variable(tf.truncated_normal(shape=(params['fc1_size'], params['fc2_size']), mean = MU, stddev = SIGMA), name='weight_4')
    fc2_b = tf.Variable(tf.zeros(params['fc2_size']), name='bias_4')
    fc2 = tf.matmul(fc1, fc2_W) + fc2_b
    fc2 = tf.nn.relu(fc2)
    fc2 = tf.nn.dropout(fc2, keep_prob=DROPOUT_RATE)
    
    # Third fully connected layer
    fc3_W = tf.Variable(tf.truncated_normal(shape=(params['fc2_size'], n_classes), mean = MU, stddev = SIGMA), name='weight_5')
    fc3_b = tf.Variable(tf.zeros((n_classes)), name='bias_5')
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return(logits)
````

````python
# Train the model
import tensorflow as tf

learning_params = [{'BATCH_SIZE': 128, 'EPOCHS': 50, 'LEARNING_RATE': 0.0004, 'MU': 0, 'SIGMA': 0.01, 'DROPOUT_RATE': 0.6},
                  {'BATCH_SIZE': 128, 'EPOCHS': 50, 'LEARNING_RATE': 0.0001, 'MU': 0, 'SIGMA': 0.01, 'DROPOUT_RATE': 0.33},
                  {'BATCH_SIZE': 32, 'EPOCHS': 50, 'LEARNING_RATE': 0.0005, 'MU': 0, 'SIGMA': 0.03, 'DROPOUT_RATE': 0.6},
                  {'BATCH_SIZE': 32, 'EPOCHS': 50, 'LEARNING_RATE': 0.0002, 'MU': 0, 'SIGMA': 0.03, 'DROPOUT_RATE': 0.9},
                  {'BATCH_SIZE': 256, 'EPOCHS': 50, 'LEARNING_RATE': 0.0004, 'MU': 0, 'SIGMA': 0.01, 'DROPOUT_RATE': 0.5}]

model_dict = [{'X_train': X_train, 'y_train': y_train,
               'save_file': 'model_color', 'num_colors': 3},
             {'X_train': X_train_gray, 'y_train': y_train,
               'save_file': 'model_gray', 'num_colors': 1},
             {'X_train': X_train_hist_eq, 'y_train': y_train,
               'save_file': 'model_histeq', 'num_colors': 3},
             {'X_train': X_train_hist_eq_gray, 'y_train': y_train,
               'save_file': 'model_histeq_gray', 'num_colors': 1},
             {'X_train': X_train_clahe_gray, 'y_train': y_train,
               'save_file': 'model_clahe', 'num_colors': 1}]

def training_pipeline(model_dict, learning_params):
    tf.reset_default_graph()
    
    LEARNING_RATE = learning_params['LEARNING_RATE']
    MU = learning_params['MU']
    SIGMA = learning_params['SIGMA']
    BATCH_SIZE = learning_params['BATCH_SIZE']
    EPOCHS = learning_params['EPOCHS']
    DROPOUT_RATE = learning_params['DROPOUT_RATE']
    num_colors = model_dict['num_colors']
    X_train = model_dict['X_train']
    y_train = model_dict['y_train']
    save_file = model_dict['save_file']
    
    params_arch = {'conv_filter_size_1': [5, 5, num_colors, 16], 
                   'conv_filter_size_2': [5, 5, 16, 32],
                   'fc1_size': 200,
                   'fc2_size': 200}
    
    # Input-output vars
    x = tf.placeholder(tf.float32, (None, 32, 32, num_colors))
    y = tf.placeholder(tf.int32, (None))
    one_hot_y = tf.one_hot(y, n_classes)

    # Logits
    logits = train_single_dnn(x, params_arch, MU, SIGMA, DROPOUT_RATE)

    # Error function
    cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
    loss_operation = tf.reduce_mean(cross_entropy)
    optimizer = tf.train.AdamOptimizer(learning_rate=LEARNING_RATE)
    training_operation = optimizer.minimize(loss_operation)
    
    # Evaluation
    correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
    accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    # Saver
    saver = tf.train.Saver()

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        num_examples = len(X_train)

        print("Training...")
        print()
        for i in range(EPOCHS):
            X_train, y_train = shuffle(X_train, y_train)
            for offset in range(0, num_examples, BATCH_SIZE):
                end = offset + BATCH_SIZE
                batch_x, batch_y = X_train[offset:end], y_train[offset:end]
                sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})

        # Save the model
        saver.save(sess, save_file)
        print('Trained Model Saved.')
        sess.close()
    tf.reset_default_graph()
        

for model in model_dict:
    for k in range(len(learning_params)):
        print(learning_params[k])
        model_copy = model.copy()
        model_copy['save_file'] = model['save_file'] + '_' + str(k) + '.ckpt'
        training_pipeline(model_copy, learning_params[k])
````

My main approach was to first split the training set further into a training set and a validation set. This splitting had to be performed at the level of **videos**, not at the level of individual images. Otherwise, this would have led to highly optimistic accuracy estimates, since the test set was from distinct videos, not distinct images. See the appendix for the code.

The architecture in Ref [2] is very similar to the LeNet architecture, but lacks the last fully connected layer. It was decided, to try the training process both with and without this layer. The cross validation results suggested that this layer was benefical. From a bagging perspective, it would actually have been preferable to even keep different network architectures, but due to a lack of time, it was decided to not pursue this approach further.

The next step was to find reasonable sets of hyperparameters (cross validation results for a single CNN and a given set of hyperparameters can be found in the files **SDNN_x.csv**): there are hyperparameter settings for which the cross validation results are not better than randomly guessing the class of the image. These setting would not have been very beneficial to the bagged model. Randomly searching over a limited set of hyperparameters (the limits were found by trial and error) is computationally less intensive than doing a grid search, but leads to similiar results [3]. The cross validation was performed for each differently preprocessed copy of the input data and the results were stored in csv files. Finally, a well performing set of five combinations of hyperparameters were chosen for the final training process.

The results are very promising with an accuracy rate of 98.3%, but they are nevertheless worse then in Ref [2]. One possible reason is that the images in [2] had a higher resolution of 48x48. Also, due to a lack of time (or knowledge) it was not possible to perform the *imadjust* preprocessing. This might have led to an additional incremental improvement.

````python
# Load pickled data
import pickle
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import numpy as np

def split_train_cv(X_train, y_train):
    train_length = X_train.shape[0] / 30
    train_ind, cv_ind = train_test_split(range(int(train_length)), test_size=0.25, random_state=0)
    
    train_ind = np.array(train_ind, dtype=int) * 30
    cv_ind = np.array(cv_ind, dtype=int) * 30
    
    training_set = np.reshape(np.array([X_train[ind:ind + 30, :, :, :] for ind in train_ind]), (-1, 32, 32, 3))
    cv_set = np.reshape(np.array([X_train[ind:ind + 30, :, :, :] for ind in cv_ind]), (-1, 32, 32, 3))
    
    training_set_y = np.reshape(np.array([y_train[ind:ind + 30:] for ind in train_ind]), -1)
    cv_set_y = np.reshape(np.array([y_train[ind:ind + 30:] for ind in cv_ind]), -1)
    
    return(training_set, training_set_y, cv_set, cv_set_y)

training_file = "./traffic-signs-data/train.p"
testing_file = "./traffic-signs-data/test.p"

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
with open(testing_file, mode='rb') as f:
    test = pickle.load(f)

X_train, y_train = train['features'], train['labels']
X_test, y_test = test['features'], test['labels']

# Split training set further into training set and validation set

X_train, y_train, X_validation, y_validation = split_train_cv(X_train, y_train)
    
assert(len(X_train) == len(y_train))
assert(len(X_validation) == len(y_validation))
assert(len(X_test) == len(y_test))

print()
print("Image Shape: {}".format(X_train[0].shape))
print()
print("Training Set:   {} samples".format(len(X_train)))
print("Validation Set: {} samples".format(len(X_validation)))
print("Test Set:       {} samples".format(len(X_test)))
````

````python
def sample_learning_params(learning_params_grid):
    sample_params = {}
    for key in learning_params_grid:
        random_index = np.random.randint(low = 0, high = len(learning_params_grid[key]))
        sample_params[key] = learning_params_grid[key][random_index]

    return(sample_params)
````

````python
import tensorflow as tf
from tensorflow.contrib.layers import flatten
import os.path
file_path = './SDNN_color.csv'
num_colors = 3

n_iterations = 30

params_learn = {'LEARNING_RATE': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005],
               'BATCH_SIZE': [16, 32, 64, 128, 256],
               'MU': [0],
               'SIGMA': [0, 0.01, 0.03],
               'EPOCHS': [50],
               'DROPOUT_RATE': [0.001, 0.25, 0.33, 0.5, 0.6, 0.75, 0.9]}

params_arch = [{'conv_filter_size_1': [5, 5, num_colors, 32], 
                'conv_filter_size_2': [5, 5, 32, 64],
                'fc1_size': 300,
                'fc2_size': 200},
               {'conv_filter_size_1': [5, 5, 1, 6],
                'conv_filter_size_2': [5, 5, 6, 16],
                'fc1_size': 300},
               {'conv_filter_size_1': [5, 5, 3, 16],
                'conv_filter_size_2': [5, 5, 16, 32],
                'fc1_size': 80},
               {'conv_filter_size_1': [7, 7, 3, 8],
                'conv_filter_size_2': [7, 7, 8, 6],
                'fc1_size': 120},
               {'conv_filter_size_1': [5, 5, 3, 32],
                'conv_filter_size_2': [5, 5, 32, 16],
                'fc1_size': 80}]

def training_pipeline(X_train, y_train, X_validation, y_validation, learning_params):
    LEARNING_RATE = learning_params['LEARNING_RATE']
    MU = learning_params['MU']
    SIGMA = learning_params['SIGMA']
    BATCH_SIZE = learning_params['BATCH_SIZE']
    EPOCHS = learning_params['EPOCHS']
    DROPOUT_RATE = learning_params['DROPOUT_RATE']
    
    # Input-output vars
    x = tf.placeholder(tf.float32, (None, 32, 32, num_colors))
    y = tf.placeholder(tf.int32, (None))
    one_hot_y = tf.one_hot(y, n_classes)

    # Logits
    logits = train_single_dnn(x, params_arch[0], MU, SIGMA, DROPOUT_RATE)

    # Error function
    cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
    loss_operation = tf.reduce_mean(cross_entropy)
    optimizer = tf.train.AdamOptimizer(learning_rate=LEARNING_RATE)
    training_operation = optimizer.minimize(loss_operation)
    
    # Evaluation
    correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
    accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    def evaluate(X_data, y_data, sess):
        num_examples = len(X_data)
        total_accuracy = 0
        for offset in range(0, num_examples, BATCH_SIZE):
            batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
            accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
            total_accuracy += (accuracy * len(batch_x))
        return total_accuracy / num_examples

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        num_examples = len(X_train)

        print("Training...")
        print()
        for i in range(EPOCHS):
            X_train, y_train = shuffle(X_train, y_train)
            for offset in range(0, num_examples, BATCH_SIZE):
                end = offset + BATCH_SIZE
                batch_x, batch_y = X_train[offset:end], y_train[offset:end]
                sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})

            validation_accuracy = evaluate(X_validation, y_validation, sess)
            print("EPOCH {} ...".format(i+1))
            print("Validation Accuracy = {:.3f}".format(validation_accuracy))
            print()
    
    learning_params['ACCURACY'] = validation_accuracy
    return(learning_params)



def tune_parameter_random_search(X_train, y_train, X_validation, y_validation, params_learn, n_iterations):
    results = []
    for k in range(n_iterations):
        params = sample_learning_params(params_learn)
        print("Learning parameters:")
        print(params)
        results_params = training_pipeline(X_train, y_train, X_validation, y_validation, params)
        if os.path.isfile(file_path):
            pd.DataFrame(results_params, index=[0]).to_csv(file_path, mode='a', header = False)
        else:
            pd.DataFrame(results_params, index=[0]).to_csv(file_path, index = False)
            
        results.append(results_params)

    return(results)

cv_results = tune_parameter_random_search(X_train, y_train, X_validation, y_validation, 
                                          params_learn, n_iterations)
````
